## Replacing null values

```python
customers["phone"]=customers["phone"].fillna("Not Available")
```

```python
customers.isnull().sum()
```

```python
# Convert to datetime
orders["shipped_date"]=pd.to_datetime(orders["shipped_date"],errors="coerce")

# Create shipping status
orders["shipping_status"]=np.where(orders["shipped_date"].isnull(),"Pending","Shipped")
```
